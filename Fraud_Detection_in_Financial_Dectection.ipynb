```python
# Importing the Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt 
```

```python
# Loading the dataset to a Pandas DataFrame credit_card
card_data = pd.read_csv('credit_data.csv')
card_data.head()
```

```python
card_data.tail()
```

```python
# dataset information
card_data.info()
```

```python
card_data.isnull().sum()
```

```python
# distribution of legit transaction & fradulent transactions
card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 -> Normal Transaction
1 -> Fradulent Transaction

```python
# Separating the data for analysis
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1] 
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
card_data.groupby('Class').mean()
```

```python
# Under-Sampling: Build a sample dataset containing similar distribution of normal and fraudulent transactions 
legit_sample = legit.sample(n=492)
# Concatenating two Dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0) 
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
# Splitting the data into Features & Targets
X = new_dataset.drop(columns='Class', axis=1)
X
```

```python
Y = new_dataset['Class'] 
Y
```

```python
# Split the data into Training Data & Testing Data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,
random_state=2) 
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model Training - Logistic Regression
model = LogisticRegression()
model.fit(X_train, Y_train) 
```

```python
# Model Evaluation - Accuracy Score
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print ('Accuracy on Training data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print ('Accuracy Score on Test data: ', test_data_accuracy)
```

```python
# Visualization using Matplotlib
accuracies = [training_data_accuracy, test_data_accuracy]
labels = ['Training Data', 'Test Data']
plt.figure(figsize=(8, 5))
plt.bar(labels, accuracies, color=['blue', 'green'])
plt.ylim(0, 1)
plt.xlabel('Data Type')
plt.ylabel('Accuracy')
plt.title('Accuracy of Training and Testing Data')
plt.show()
```
